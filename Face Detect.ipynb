Loading the classifier for frontal face

```python
haar_cascade_face =cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
# 'Haar cascade' is one of the predefined templates used for object detection.
# "haarcascade_frontalface_default.xml" is a predefined model for face detection.
```

Suppose we are reading a human face, the portion of area of eyes should be darker and above that darker the forehead portion should be brighter and again below the darker area there will be a bright area. This is one of the pattern is used in face detection.



This type of pattern is called haarcascade. From the image we have to find out this kind of area or pattern. Like this we also have haarcascade for mask detection. The different types of pattern for object detection as shown below.

Below image shows different types of pattern used to detect different portion of the face.


### Face detection

```python
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5);

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

```python
# Creating square to detect the face. 
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (255, 0, 0), 50)
```

```python
# Face Detection with generalized function
```

```python
def detect_faces(cascade, test_image, scaleFactor = 1.1):
    # create a copy of the image to prevent any changes to the original one.
    image_copy = test_image.copy()

    #convert the test image to gray scale as opencv face detector expects gray images
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

    # Applying the haar classifier to detect faces
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=5)

    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)

    return image_copy
```
